## Sample Data Generation for Snacks Recommendation

We will create 2 different datasets using Faker and python using some Kaggle datasets too. 

#### 1. CTR Click Prediction data - 
This dataset is based on Cretio Click through rate dataset. We have columns like Customer name, Campaign details with categorical features related to the ad shown to users, user details and whether they clicked an Advertisement for a product or not. This will be help us in recommending Products/Campaigns/Ads to customers that they are more likely to click and convert.
##### We will use this dataset to demonstrate the xDeepFM Recommendation System.


#### 2. Historical Purchase Data - 
This data is based on Instacart market basket data. It has information like product details, departments etc. Later we create the list of user journey in terms of orders made by a user in chronological order. This will help us in recommending products to customer which they are most likely to add to their cart or just continue their subscription for a particular snack category. It can also help us in retaining customers who can churn in future. 
##### We will use this dataset to demonstrate the Customer2Vec Recommendation System. 

```python
import pandas as pd
import os 
import numpy as np
```

```python
# reading criteo_sample data
os.chdir(r'N:\ALGORITHMIC MARKETING\Assignment4')

data = pd.read_csv('data/criteo_sample.txt')
data.head(10)

np.random.seed(42)
```

```python
data.describe()
```

```python
data.describe(include=['object'])
```

```python
data['Uni_id1'] = data['C1']
data['Uni_id2'] = data['C23']

uni1 = data.C1.unique()
uni2 = data.C23.unique()
```

```python
data.head()
```

```python
from faker import Faker
import pandas as pd

faker = Faker()

df_Customer = []

for n in range(27):
    df_Customer.append({'Latitude': faker.coordinate(center=74.0, radius=0.10),
               'Longitude': faker.coordinate(center=40.8, radius=0.10),
               'Text': faker.sentence(),
               'Customer_Name': faker.name(),
               'Address': faker.address(),
               'Job': faker.job()
              })

df_Customer = pd.DataFrame(df_Customer)
df_Customer = df_Customer[['Customer_Name', 'Job', 'Text', 'Address', 'Latitude', 'Longitude']]
df_Customer.head(10)
```

```python
df_Customer['Customer_Id'] = np.arange(len(df_Customer))
df_Customer.head()
```

```python
df_Customer['Uni_id1'] = uni1
```

```python
# Product Campaign Data

Campaign_name = ['Snakku', 'Love with food', 'Candy Club', 'Nature Box', 'ZenPop', 'World Sampler', 'FitSnack', 'Vegan Cuts',
                'MunchPak', 'KetoKrate']

Campaign_id = np.arange(100230, 100240)
```

```python
df_campaign = pd.DataFrame()

df_campaign['Campaign_name'] = Campaign_name
df_campaign['Campaign_id'] = Campaign_id
```

```python
df_campaign['Uni_id2'] = uni2
df_campaign
```

```python
df_Snacks_CTR = pd.merge(data, df_Customer[['Uni_id1', 'Customer_Name', 'Customer_Id']],
                    on='Uni_id1', how='left')
df_Snacks_CTR
```

```python
df_Snacks_CTR = pd.merge(df_Snacks_CTR, df_campaign[['Uni_id2', 'Campaign_name', 'Campaign_id']],
                    on='Uni_id2', how='left')
df_Snacks_CTR
```

```python
df_Snacks_CTR.columns
```

```python
df_Snacks_CTR = df_Snacks_CTR.drop(columns=['Uni_id1', 'Uni_id2'])
df_Snacks_CTR
```

```python
df_Snacks_CTR.to_csv('Snacks_CTR_Data.csv')
```
